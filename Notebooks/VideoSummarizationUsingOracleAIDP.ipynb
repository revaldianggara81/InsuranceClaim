## Create a summary of the key events depicted in the "Tailgate" video

```python
import os, io, base64, cv2
from PIL import Image
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

def encode_image_b64(img_bgr, max_side=1280, jpeg_quality=85):
    h, w = img_bgr.shape[:2]
    scale = min(1.0, float(max_side) / max(h, w))
    if scale < 1.0:
        img_bgr = cv2.resize(img_bgr, (int(w*scale), int(h*scale)), interpolation=cv2.INTER_AREA)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    buf = io.BytesIO()
    Image.fromarray(img_rgb).save(buf, format="JPEG", quality=jpeg_quality, optimize=True)
    b64 = base64.b64encode(buf.getvalue()).decode("utf-8")
    return f"data:image/jpeg;base64,{b64}"

def sample_frames(video_path, max_frames=12, min_interval_s=0.5):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise RuntimeError(f"Cannot open video: {video_path}")
    fps = cap.get(cv2.CAP_PROP_FPS) or 24.0
    nframes = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    dur = nframes / max(fps, 0.0001)

    if dur <= 0:
        indices = list(range(min(nframes, max_frames)))
    else:
        count = min(max_frames, max(1, int(dur / min_interval_s)))
        indices = [int(round(i * (nframes - 1) / (count - 1))) for i in range(count)] if count > 1 else [0]

    imgs, stamps = [], []
    for idx in sorted(set(indices)):
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ok, frame = cap.read()
        if not ok:
            continue
        imgs.append(frame)
        stamps.append(idx / max(fps, 1e-6))
    cap.release()
    return imgs, stamps, dur

def build_multimodal_messages(encoded_images, stamps, instructions):
    content = [{"type": "text", "text": instructions}]
    for img, t in zip(encoded_images, stamps):
        content.append({"type": "text", "text": f"[t={round(t, 2)}s]"})
        content.append({"type": "image_url", "image_url": {"url": img}})
    return [
        {"role": "system", "content": "You are a meticulous video analyst. Describe only what is visually evident."},
        {"role": "user", "content": content}
    ]

# Ollama on-premises client
client = OpenAI(
    base_url=os.getenv("OLLAMA_BASE_URL", "http://localhost:11434/v1"),
    api_key="ollama"
)
VLM_MODEL = os.getenv("OLLAMA_VLM_MODEL", "qwen2.5vl:7b")

print("Sampling frames…")
frames, stamps, dur = sample_frames("/uploads/tailgate.MP4", max_frames=12)

if not frames:
    raise RuntimeError("No frames sampled; check the video file.")

print(f"Encoding {len(frames)} frames…")
encoded = [encode_image_b64(f) for f in frames]

instructions = f"""
You will analyze still frames sampled from a single video with NO AUDIO.
Task:
1) Produce a short overview (2–4 sentences) of the visible events.
2) Then produce a chronological TIMELINE with bullet points: [mm:ss] what happens (verbs, people, actions, objects, entrances/exits).
3) Be explicit about access control / tailgating if suspected, but do not guess identities, intent, or unseen causes.
4) Use only visual evidence; avoid speculation. The video duration ~{dur:.1f}s.
Output sections:
== Overview ==
== Timeline ==
"""

messages = build_multimodal_messages(encoded, stamps, instructions)

print(f"Calling {VLM_MODEL} via Ollama…")
resp = client.chat.completions.create(
    model=VLM_MODEL,
    messages=messages
)

print("\n=== SUMMARY ===\n")
print(resp.choices[0].message.content)
```

## Store the event summary from the "Tailgate" video in a table within the "Silver" layer

```python
from pyspark.sql import SparkSession, Row
from pyspark.sql.functions import lit

my_string = resp.choices[0].message.content
data = [("/Volumes/demo/bronze/video_input/tailgate.MP4",my_string,)]
columns = ["video_path","detailed_summary"]
df = spark.createDataFrame(data, columns)

spark.sql("create table if not exists demo.silver.video_summary (video_path string, detail_summary string) using delta").show(truncate=False)

df.write.mode("overwrite").insertInto("demo.silver.video_summary")

spark.sql("select * from demo.silver.video_summary").show( 10, False)
```

## Use case# 2: Let's examine a brief video from the "Bronze" layer that demonstrates a "Car crash" incident

```python
import io
import base64
from IPython.display import HTML

def display_video(path):
  mp4 = io.open(path, 'r+b').read()
  data_url = "data:video/mp4;base64," + base64.b64encode(mp4).decode()
  return HTML(f"""
      <video width=400 controls>
          <source src="{data_url}" type="video/mp4">
      </video>
  """)

display_video("/Volumes/demo/bronze/video_input/CarCollision.mp4")
```

## Create a summary of the key events depicted in the "Car crash" video

```python
import os, io, base64, cv2
from PIL import Image
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

# Reuse helper functions from cell above (encode_image_b64, sample_frames, build_multimodal_messages)

client = OpenAI(
    base_url=os.getenv("OLLAMA_BASE_URL", "http://localhost:11434/v1"),
    api_key="ollama"
)
VLM_MODEL = os.getenv("OLLAMA_VLM_MODEL", "qwen2.5vl:7b")

print("Sampling frames…")
frames, stamps, dur = sample_frames("/Volumes/claims_bronze/claims_db/evidence/claimant_video_evidence/CarCollision.mp4", max_frames=63)

if not frames:
    raise RuntimeError("No frames sampled; check the video file.")

print(f"Encoding {len(frames)} frames…")
encoded = [encode_image_b64(f) for f in frames]

instructions = f"""
You will analyze still frames sampled from a single traffic or car accident video with NO AUDIO.
Task:
1) Produce a short overview (2–4 sentences) describing the accident: the vehicles involved, the sequence of impact, and the general scene.
2) Then produce a chronological TIMELINE with bullet points: [mm:ss] what happens (vehicle movements, collisions, road features, pedestrians, bystanders, emergency responses).
3) Be explicit about vehicle interactions (e.g., collision, near-miss, sudden braking, swerving), but do not speculate about causes, driver intent, or fault.
4) Use only visible evidence; avoid assumptions about unseen factors. The video duration is ~{dur:.1f}s.
Output sections:
== Overview ==
== Timeline ==
"""

messages = build_multimodal_messages(encoded, stamps, instructions)

print(f"Calling {VLM_MODEL} via Ollama…")
resp = client.chat.completions.create(
    model=VLM_MODEL,
    messages=messages
)

print("\n=== SUMMARY ===\n")
print(resp.choices[0].message.content)
```

## Store the event summary from the "Car crash" video in a table within the "Silver" layer

```python
from pyspark.sql import SparkSession, Row
from pyspark.sql.functions import lit

my_string = resp.choices[0].message.content
data = [("/Volumes/demo/bronze/video_input/CarCollision.mp4",my_string,)]
columns = ["video_path","detailed_summary"]
df = spark.createDataFrame(data, columns)

df.write.mode("append").insertInto("demo.silver.video_summary")

spark.sql("select * from demo.silver.video_summary").show( 10, False)
```

## Next, let's perform sentiment analysis on each detailed summary stored in the "Silver" layer and save the resulting sentiment data in the "Gold" layer

```python
from pyspark.sql.functions import expr

df = spark.read.table("demo.silver.video_summary")

df = df.withColumn(
    "is_car_crash",
    expr(
        "query_model('xai.grok-3-fast', concat('Answer should return only positive or negative. Is this a car crash?',detail_summary))"
    )
)
df2 = df.withColumn(
    "is_tailgate",
    expr(
        "query_model('xai.grok-3-fast', concat('Answer should return only positive or negative. Is this a tailgate?',detail_summary))"
    )
)
spark.sql("create table if not exists demo.gold.video_dataset(video_path string, detail_summary string, is_car_crash string, is_tailgate string) using delta").show(truncate=False)

df2.write.mode("overwrite").insertInto("demo.gold.video_dataset")

```

```python
df3 = spark.sql("select * from demo.gold.video_dataset")
df3.createOrReplaceTempView("video_dataset")
```
